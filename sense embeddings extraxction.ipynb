```python
# getting list of all prepared sentences
def get_dev_text():
    dev_text = open('/home/sara/data_file_dev.text', "r")
    dev_text = dev_text.readlines()

    dev_text1 = []
    for i in dev_text:
        dev_text1.append(i.replace('\n', ''))
    return dev_text1


#getting list of all possible senses of sense word sentences
def get_senses():
    output_vocab = open('/media/sara/Data1/data/output_data/output_vocabulary0', "r")
    output_vocab1 = output_vocab.readlines()
    output_vocab2 = []
    for i in output_vocab1:
        output_vocab2.append(i.replace('\n', ''))
    
    dev = open("/media/sara/Data1/data/output_data/dev", "r")
    dev = dev.readlines()
    
    #lists of indices of sentences, senses and multiple senses
    sentence_dev_indices = dev[0::3]
    sense_dev_indices = dev[1::3]
    multiple_sense_dev_indices = dev[2::3]

    multiple_sense_dev_indices 
    multiple_sense_dev_indices1 = [i.replace('\n', '') for i in multiple_sense_dev_indices] # removing \n

    multiple_sense_dev_indices2= [x.split(' ')for x in multiple_sense_dev_indices1] 


    a = []
    for sentence in multiple_sense_dev_indices2:
        sense_position_new = []
        for sense_position in sentence:
            if sense_position == '':
                continue
            sense_position_new.append(sense_position.split(';'))
        a.append(sense_position_new)

    senses = []
    for sentence in a:
        senses1 = []
        for sense_position in sentence:
            senses2 = []
            for sense_index in sense_position:
                try:
                    sense_index_int = int(sense_index)
                    sense = output_vocab2[sense_index_int]
                    senses2.append(sense)
                except Exception:
                    print(f'sense_index: {sense_index}')
                    print(f'sense_position: {sense_position}')
                    print(f'sentence: {sentence}')
                    raise Exception
            senses1.append(senses2)
        senses.append(senses1)
    return senses
```

```python
#Masking senses having multiple senses in sentence

def convert_list_to_string(org_list, seperator=' '):
    """ Convert list to string, by joining all item in list with given separator.
        Returns the concatenated string """
    return seperator.join(org_list)

def data_preparation(sentence, sentence_senses):
    
    sentence_splitted = sentence.split(' ')

    multisense_sentence = []   #multisense senses of the sentence
    multi_sense_index = []  # index of sense (having multiple senses) in senses
    for index, element in enumerate(sentence_senses):
        if len(element) > 1:
            multi_sense_index.append(index)
            multisense_sentence.append(element)
    #to get correct sense
    index_sent = [] #index of multisenses senses in sentence
    for i in multi_sense_index:
        ind_sent  = i*2
        index_sent.append(ind_sent)
    correct_sense = []
    for index, element in enumerate(sentence_splitted):
        for ind in index_sent:

            if index == ind:
                correct_sense.append(element)

    sentence_list = sentence.split(' ') #converting sentence string into list of strings of indivudual words
    #masking multisenses sense            
    sentence_list = sentence.split(' ')
    for ind in index_sent:
        sentence_list[ind] = '[MASK]'
    final_sentence = convert_list_to_string(sentence_list)
    return final_sentence, multisense_sentence, correct_sense
```

```python
def model_prediction(final_sentence, tokenizer, model):
    idp=tokenizer(final_sentence, return_tensors="pt" ,add_special_tokens=True, truncation=True, max_length=512) ["input_ids"]
    output = model(idp, labels=idp, output_hidden_states=True)
    loss, prediction_scores = output[:2]
    output_hidden_states = output[-1]
    #print(loss, prediction_scores)
    return prediction_scores, loss, output_hidden_states
```

```python
import numpy as np

def result_comparison(final_sent, correct_sense, prediction_scores, multisense_sentence, tokenizer):

    tokenized_final_sentence = np.array([tokenizer.ids_to_tokens[idx] for idx in tokenizer.encode(final_sent)])
    mask_indices = np.where(tokenized_final_sentence == '[MASK]')[0] 
    
    correct = 0
    incorrect = 0
    for j in range(len(correct_sense)):
        tokenizer_id = []
        for i in multisense_sentence[j]:
            tokenizer_id.append(tokenizer.vocab[i])


        pred_scor = prediction_scores[0,mask_indices[j], tokenizer_id]
        sense_masked1  = multisense_sentence[j] # list of multiple senses of first masked sense in the sentence


        eval1 = sense_masked1 [pred_scor.argmax()] #maximum prediction score among senses
        if eval1 == correct_sense[j]:
            correct += 1
        else:
            incorrect += 1
    return correct, incorrect
```

```python
from transformers import BertForMaskedLM, BertTokenizer
def prelimanaries():
    dev_text1 = get_dev_text()
    senses = get_senses()
    tokenizer = BertTokenizer.from_pretrained('/media/sara/Data1/data/bert_training_l2l_256block/')
    model = BertForMaskedLM.from_pretrained('/media/sara/Data1/data/bert_training_l2l_256block/checkpoint-2500/')
    total_correct  = 0
    total_incorrect = 0
    for l in range(len(dev_text1)):
        sentence = dev_text1[l]
        sentence_senses = senses[l]
        final_sent, multisense_sentence, correct_sense = data_preparation(sentence, sentence_senses)
        prediction_scores, loss, output_hidden_states= model_prediction(final_sent, tokenizer, model)
        correct, incorrect = result_comparison(final_sent, correct_sense, prediction_scores, multisense_sentence, tokenizer)
        total_correct += correct
        total_incorrect += incorrect
    return total_correct, total_incorrect, total_correct/(total_correct+total_incorrect)
```
