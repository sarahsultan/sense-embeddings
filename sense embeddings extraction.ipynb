```python
#data_semcor_wng
output_vocab_semwng = open('/media/sara/Data1/data/output_data_senseval/output_vocabulary0', "r")
output_vocab_semwng = output_vocab_semwng.readlines()

train_semwng = open("/media/sara/Data1/data/output_data_senseval/train", "r")
train_semwng = train_semwng.readlines()

input_vocab_semwng = open('/media/sara/Data1/data/output_data_senseval/input_vocabulary0', "r")
input_vocab_semwng = input_vocab_semwng.readlines()

#lists of indices of sentences, senses and multiple senses
sentence_indices = train_semwng[0::3]
sense_indices = train_semwng[1::3]
multiple_sense_indices = train_semwng[2::3]
```

```python
def convert_string_to_list(string): 
        lis = list(string.split(" ")) 
        return lis
def convert_list_to_string(org_list, seperator=' '):
    """ Convert list to string, by joining all item in list with given separator.
        Returns the concatenated string """
    return seperator.join(org_list)

def sense_and_sentence(string_of_isentence, string_of_isense, input_vocabulary, output_vocabulary):
    # getting a single sentence
    splitted_sen = string_of_isentence.replace('\n','')
    #print(splitted_sen)
    splitted_sen = splitted_sen.split(' ')
    #print(splitted_sen)
    int_sen = [int(x) for x in splitted_sen if x != '']
    word_list = []
    for a in int_sen:
        word_list.append(input_vocabulary[a])
    one_sen = ' '.join([g for g in word_list])
    final_sen = one_sen.replace('\n', '')
   
    #getting a single sense
    splitted_sen1 = string_of_isense.replace('\n','')
    splitted_sen1 = splitted_sen1.split()
    int_sen1 = [int(x) for x in splitted_sen1 if x!= '']
    word_list1 = []
    for a in int_sen1:
        word_list1.append(output_vocabulary[a])

    one_sen1 = ' '.join([g for g in word_list1])
    final_sen1 = one_sen1.replace('\n', '')
    final_sen1

    #converting strings of sense and sentence to lists

    sentence = convert_string_to_list(final_sen)
    sense = convert_string_to_list(final_sen1)
    # merging sense and sentence
    list1 = sense
    list2 = sentence
    result = [None]*(len(sense)+len(sentence))
    result[::2] = sense
    result[1::2] = sentence
    result1 = convert_list_to_string(result)
    return result1

```

```python
final_list_semwng = []
for (elem_isentence, elem_isense) in zip(sentence_indices, sense_indices):
    result = sense_and_sentence(elem_isentence, elem_isense, input_vocabulary=input_vocab_semwng, output_vocabulary=output_vocab_semwng)
    final_list_semwng.append(result)
```

```python
final_list_semwng = [line+'\n' for line in final_list_semwng]
```
